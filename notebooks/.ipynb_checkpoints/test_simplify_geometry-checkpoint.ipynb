```python
# utility functions 

def print_geoms_name(geom, init_buf=""):
    if geom.GetGeometryName() == "GEOMETRYCOLLECTION":
        print("%s  * GEOMETRYCOLLECTION" % (init_buf))
        for i in range(0, geom.GetGeometryCount()):
            print_geoms_name(geom.GetGeometryRef(i), init_buf + "  ")
    else:
        print("%s  * %s" % (init_buf, geom.GetGeometryName()))


def print_geom_name(geom):
    """
        input:      geom = ogr.Geomtry
        output:     a string like GEOMETRYCOLLECTION ( POLYGON(), LINESTRING(), ...)
    """
    if geom.GetGeometryCount() == 0:
        return ""
    if geom.GetGeometryCount() > 1 or \
        geom.GetGeometryName() in ["GEOMETRYCOLLECTION",
                                   "MULTIPOINT", "MULTIPOLYGON",
                                   "MULTILINESTRING"]:
        s = geom.GetGeometryName()
        s += " ( "
        for i in range(0, geom.GetGeometryCount()):
            s += print_geom_name(geom.GetGeometryRef(i)) + ", "
        s = s[:len(s)-2]
        s += " )"
        return s
    return geom.GetGeometryName()

        
def print_geoms_wkt(geom, init_buf=""):
    if geom.GetGeometryName() == "GEOMETRYCOLLECTION":
        print("%s  * GEOMETRYCOLLECTION" % (init_buf))
        for i in range(0, geom.GetGeometryCount()):
            print_geoms_wkt(geom.GetGeometryRef(i), init_buf + "  ")
    else:
        print("%s  * %s" % (init_buf, geom.ExportToWkt()))


def print_simple_geoms(geom, init_buf=""):
    cnt = geom.GetGeometryCount()
    if cnt > 0:
        print("%s%s has %d shapes inside" % (init_buf, geom.GetGeometryName(), cnt))
        for i in range(0, cnt):
            gref = geom.GetGeometryRef(i)
            print_simple_geoms(gref, init_buf + "  ")
    else:
        print("%s%s" % (init_buf, geom.GetGeometryName()))



def simplify_prism_geometry(mgeom):
    """
        input:               mgeom = ogr.GeometryCollection, ogr.Multipolygon or ogr.Polygon
                                      containing multiple shapes inside the first one of 
                                      which will be used to create a polygon
        output:              ogr.Geometry(Polygon) or None (if error)
    """
    try:
        print("wotking with %s" % mgeom.GetGeometryName())
        # special cases: empty geometry, polygon or geometrycollection
        if mgeom.GetGeometryCount() == 0:
            return None
        if mgeom.GetGeometryName() == "POLYGON":
            return mgeom.Clone()
        if mgeom.GetGeometryName() == "GEOMETRYCOLLECTION":
            return simplify_prism_geometry(mgeom.GetGeometryRef(0))
        # normal execution: multipolygon case
        if mgeom.GetGeometryName() == "MULTIPOLYGON":
            poly = ogr.Geometry(ogr.wkbPolygon)
            rf = poly
            for i in range(0, mgeom.GetGeometryCount()):
                poly = poly.Union(mgeom.GetGeometryRef(i))
                print("    successful union" if rf == poly else "    unsuccessful union")
                rf = poly
            return poly
    except Exception as e:
        print("[simplify_prism_geometry] exception: %s" % e)
        None
    print("[simplify_prism_geometry]: unsupported input %s" % mgeom)
    return None



def simplified_roadgraph_from_xml(xmltree_roadgraph):
    rgraph = ogr.Geometry(ogr.wkbGeometryCollection)
    err_list = list() # pair of original WKT and output ogr.Polygon
    segments = [s for s in xmltree_roadgraph.find("segments")]
    print("* found %d segment(s) in roadgraph(%d)" % (len(segments), nlev))
    for seg in segments:
        wktShape = seg.find("geometry").find("polygon").text
        gcoll = ogr.CreateGeometryFromWkt(wktShape)
        poly = simplify_prism_geometry(gcoll)
        if poly is None or poly.GetGeometryCount() == 0:
            err_list.append((gcoll, poly))
        rgraph.AddGeometry(poly)
    print("* completed road graph geometry simplification with %d errors (on %d cases)" % (
            len(err_list), len(grd_segs)))
    #print("* here it is:\n\n%s\n\n" % rgraph.ExportToWkt())
    return rgraph



def get_total_bounding_box(geom):
    """
        input:       geom = ogr.Geometry
        output:      tuple (pmin, pmax) in 3D obtained across all points of all shapes of geom
    """
    # Using GDAL's boundingbox "env3d = pgeom.GetEnvelope3D()" on 
    #     the parking space GeometryCollection gave incorrect points
    #     so I had to manually extract all points of all shapes composing the parking, 
    #     taking the min/max and create my own bounding box 3d.
    # After this, I created a 2d polygon out of it
    if geom.GetGeometryCount() == 0:
        return None, None
    min_max = [None, None] # (x,y,z) min + (x,y,z) max
    stack = list()
    stack.insert(0, geom)
    while len(stack) > 0:
        g = stack.pop(0)
        if g.GetGeometryCount() > 0 or \
            g.GetGeometryName() in ["GEOMETRYCOLLECTION",
                                   "MULTIPOINT", "MULTIPOLYGON",
                                   "MULTILINESTRING"]:
            for i in range(0, g.GetGeometryCount()):
                stack.insert(0, g.GetGeometryRef(i))
        else:
            for pt in g.GetPoints():
                if min_max[0] is None or pt < min_max[0]:
                    min_max[0] = pt
                if min_max[1] is None or pt > min_max[1]:
                    min_max[1] = pt
    return (min_max[0], min_max[1])


def simplify_parking_geometry(pgeom):
    """
        input:        pgeom = ogr.GeometryCollection of the DXF of a Parking Space
        outpu:        ogr.Polygon (3D)
    """
    env3d = get_total_bounding_box(pgeom)
    if env3d == (None, None):
        return None
    pmin = env3d[0]
    pmax = env3d[1]

    ring = ogr.Geometry(ogr.wkbLinearRing)
    ring.AddPoint(pmin[0], pmin[1], pmin[2])
    ring.AddPoint(pmax[0], pmin[1], pmin[2])
    ring.AddPoint(pmax[0], pmax[1], pmin[2])
    ring.AddPoint(pmin[0], pmax[1], pmin[2]) # stays on the floo
    ring.AddPoint(pmin[0], pmin[1], pmin[2])
    
    poly = ogr.Geometry(ogr.wkbPolygon)
    poly.AddGeometry(ring)
    return poly
```

```python
# to see how the function is done inside cad2xml
def get_entity_polyface(entity, verbose=False):
    facepoly = ogr.Geometry(ogr.wkbMultiPolygon)
    for sf in entity:
        ring = ogr.Geometry(ogr.wkbLinearRing)
        indices = sf.indices()
        # add all points 
        for idx in indices:
            pts = entity.vertices[idx].location
            dlib.add_point_to_shape(ring, pts, verbose)
        # add back the first point, to close the ring
        if len(indices):
            pts = entity.vertices[indices[0]].location
            dlib.add_point_to_shape(ring, pts, verbose)
        subpoly = ogr.Geometry(ogr.wkbPolygon)
        subpoly.AddGeometry(ring)
        if verbose:
            print("        ... sub-face %s" % ("Valid" if subpoly.IsValid() else "Invalid") )
        facepoly.AddGeometry(subpoly)
        if verbose:
            print("    ... face %s" % ("Valid" if facepoly.IsValid() else "Invalid") )
    return facepoly
```
